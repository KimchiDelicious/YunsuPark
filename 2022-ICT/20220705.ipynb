```python
class Cal:
    result = 0
    
    def plus(self, num):
        self.result += num
        return self.result
    
    def minus(self, num):
        self.result -= num
        return self.result
    
    def divide(self, num):
        self.result /= num
        return self.result
    
class Cal_change(Cal):
    def mul(self, num):
        self.result *= num
        return self.result
    def clear(self):
        result=0
        return self.result

c1 = Cal_change()
c2 = Cal_change()
c3 = Cal_change()
c4 = Cal_change()
c5 = Cal_change()
```
