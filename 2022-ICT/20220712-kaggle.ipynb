```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
train = pd.read_csv("../input/bikemod/bike_mod_tr.csv", parse_dates=['datetime'])
test = pd.read_csv("../input/bikemod/bike_mod_test.csv", parse_dates=['datetime'])
sub = pd.read_csv("../input/bike-sharing-demand/sampleSubmission.csv")
```

```python
print(train.info())
print()
print(test.info())
```

```python
input_col = ['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed', 'dayofweek']
labeled_col = ['count']
```

```python
X = train[ input_col ]
y = train[ labeled_col ]
X_val = test[input_col]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                random_state=0)
```

```python
### 난수 발생 패턴 결정 0
seed = 0
np.random.seed(seed)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(30, input_dim=9, activation='relu'))
model.add(Dense(15, activation='relu') )
model.add(Dense(15, activation='relu') )
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss = 'mean_squared_error', optimizer='rmsprop')
model.fit(X_train, y_train, epochs=20, batch_size=10)
```

```python
model.evaluate(X, y)
```

```python
pred = model.predict(X_val)
```

```python
sub['count'] = pred
sub.loc[sub['count']<0, 'count'] = 0
```

```python
sub.to_csv("submission.csv", index=False)
```
