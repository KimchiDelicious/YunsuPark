```python
### plot, bar 그래프 그리기
x = np.arange(0.0, 2*np.pi, 0.1)
sinx = np.sin(x)
cosx = np.cos(x)
tanx = np.tan(x)
artanx= np.arctan(x)

fig, axs = plt.subplots(2, 2, figsize=(10,10))
axs[0,0].plot(x, sinx)
axs[0,0].set_title("sin")

axs[0,1].plot(x, cosx, 'rs')
axs[0,1].set_title('cos')

axs[1,0].plot(x, tanx, 'c.')
axs[1,0].set_title('tan')

axs[1,1].plot(x, artanx, 'yo')
axs[1,1].set_title('arctan')

plt.show()
```

```python
x = np.arange(10)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.plot(x, 2*x, label='y = %ix' % 2)
plt.legend()
plt.show()
```

```python
for i in range(5):
    plt.plot(x, i*x, label='y = %ix' % i)
plt.legend()
plt.show()
```

### 실습

```python
x = train['count']
y = train['temp']

plt.plot(x,y,'o',alpha=0.2)
plt.xlabel('count')
plt.ylabel('temp')
```

```python
corr = train.corr()
corr['count']
```

```python
dic=sorted(dict(corr['count']).items(), key=lambda x:x[1], reverse=True)
dic=dict(dic)
x=dic.keys()
y=dic.values()
fig=plt.figure(figsize=(20,5))
plt.bar(x,y)
plt.ylabel('count')
plt.title('correlations with count')

plt.show()
```
